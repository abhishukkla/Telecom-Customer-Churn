```python
def conv(value):
  if value ==1:
    return 'Yes'
  else:
    return 'no'
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.title('Count of customers by churn')
plt.show()
```

```python
gb = df.groupby("Churn").agg({"Churn":"count"})
#gb
plt.pie(gb['Churn'],labels=gb.index,autopct='%1.1f%%')
plt.title("percentage of churned customers")
plt.show()
```

```python
sns.countplot(x='gender', data=df,hue="Churn")
plt.title("Churn by gender")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x = 'SeniorCitizen', data = df, hue = 'Churn')
plt.title('Churn by SeniorCitizen')
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x = 'SeniorCitizen', data = df)
plt.title('Count of SeniorCitizen')
plt.show()
```

```python
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Normalize to percentages
percent_data = count_data.div(count_data.sum(axis=1), axis=0) * 100

# Plot
ax = percent_data.plot(kind='bar', stacked=True, figsize=(4, 4), colormap='Pastel1')

# Add percentage labels
for i, (index, row) in enumerate(percent_data.iterrows()):
    cumulative = 0
    for j, val in enumerate(row):
        plt.text(i, cumulative + val / 2, f'{val:.1f}%', ha='center', va='center', fontsize=9)
        cumulative += val

# Title and labels
plt.title('Churn by SeniorCitizen (Percentage)')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'], rotation=0)
plt.legend(title='Churn', bbox_to_anchor=(0.9,0.9))
plt.tight_layout()
plt.show()
```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up subplots (3 rows x 3 cols)
fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.flatten()

# Plot each countplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, hue='Churn', ax=axes[i])
    axes[i].set_title(f'Churn by {col}')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-labels if needed

# Tidy layout
plt.tight_layout()
plt.show()
```

Customers with Fiber optic internet show the highest churn, while those with DSL or no internet churn less. Having security-related services like OnlineSecurity, OnlineBackup, DeviceProtection, and TechSupport is strongly linked to lower churn rates. In contrast, customers with MultipleLines or only entertainment services like StreamingTV/Movies are more likely to churn. Overall, value-added services and support features appear to reduce churn significantl
